# Setting Index to release Date

```python
sf.set_index("release_date" , inplace = True)
sf.index = pd.to_datetime(sf.index)
sf.head()
```

# Converting songs duration from Milliseconds to Seconds

```python
sf["duration"]=sf["duration_ms"].apply(lambda x: round(x/1000))
sf.drop("duration_ms" , inplace = True , axis=1)
```

```python
# converted Milliseconds to Seconds

sf.duration.head() 
```

# Visualising Correlation Map

```python

corr_sf1 = sf.drop(["key","mode","explicit"] , axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_sf1 ,annot= True , fmt=".1g" , vmin= -1 , vmax = 1 , center=0 , cmap="BrBG" , linecolor ="Black" , linewidths=1)
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

-1 indicates variables have least or negative correlation 
while values which are above 0.00 have positive correlation

# Creating sample data by using original Dat

```python
sample_sf = sf.sample(int(0.004*len(sf)))
```

```python
print(len(sample_sf))
```

# Regression plot between Loudness and Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_sf , y = "loudness" , x = "energy" , color = "violet" ,line_kws={"color": "black"}).set(title="Loudness vs Energy Correlation")
```

# Regression plot between Popularity and Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_sf , y="popularity" , x = "acousticness" , color = "c" , line_kws={"color": "black"} ).set(title = "popularity vs acousticness")
```

# Regression plot between Speechiness and Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_sf, y='speechiness', x='acousticness', scatter_kws={"color": "goldenrod"}, line_kws={"color": "black"}).set(title="Speechiness Vs acousticness")
```

# Extracting years from "dates" column and creating a new "years" column

```python
sf['dates']=sf.index.get_level_values('release_date')
sf.dates= pd.to_datetime(sf.dates)
years = sf.dates.dt.year
```

# Creating distribution plot to visualise total no of songs in each year since 1922 in spotify

```python
sns.displot(years , discrete=True , aspect=2 , height=5 , kind="hist").set(title = "Number of songs per year")
```

Result - Increasingh popularity of Music indicates that there were few songs available in past years but because of music app like Spotify in earlier years the no of listeners also no of songs have increased exponentially 

# Creating Barplot for duration of songs

```python
 total_dr = sf.duration
fig_size = (18 , 8)
fig , ax = plt.subplots(figsize = fig_size)
fig = sns.barplot( x = years , y = total_dr , ax = ax ,errwidth = False ).set(title = "Year vs Duration")
plt.xticks(rotation = 90 )
```
